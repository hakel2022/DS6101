### 2.	Compare Anaconda with pip/virtualenv for managing environments and packages.

Anaconda is a package that includes python, data science libraries (e.g. Pandas, NumPy), graphical user interface (Anaconda navigator) and an interactive coding tool (Jupyter Notebook).  
Its main advantage is that it offers pre-installed packages which are optimized for data science and reduce compatibility issues. The GUI also facilitates the environments management. On the other hand the main disadvantage is that it requires large disk space usage and package update is slower compared to Python online repository.  

On the other hand, pip is Python’s package manager and virtualenv creates isolated Python environments. Their main advantage is that they use minimal disk space, and they make latest package accessible. They are also flexible to accommodate custom projects beyond data science. Their main disadvantage is that they require manual setup for dependencies and command-line proficiency since there is no GUI.


### 3.	Why is Version Control Important in Collaborative Data Science Projects?

```python
Version control is important in collaborative data science mainly because of the following:  
•	Track changes: It records who changed what and when. It allows rollback if errors occur and facilitates debugging.  
•	Collaboration: It allows multiple team members to work on code simultaneously without conflicts.  
•	Version control also supports experimenting new features without affecting the main codebase. The new features can then be pushed to main codebase once they pass the test.
```

### 4.	What are the key differences between Jupyter Notebook and JupyterLab

```python
The main difference between Jupyter Notebook and JupyterLab is that Jupyter Notebook is limited to one notebook at a time whereas JupyerLab provides advance file management, supporting multiple tables and navigating datasets or outputs.
Hence, Jupyter Notebook is simple, beginner-friendly for single task whereas JupyterLab is suited for complex, multi-file workflows.
For instance, the Jupyter Notebook is appropriate for a data exploration task of a CSV dataset. On the other hand, if I need to build a machine learning model using multiple files, I would need to use JupyterLab.
```

### 5.	How Git and GitHub Work Together. What is a Pull Request and When is it Used?

```python
Git is a local version control system tracking changes in files, commits and merges on the local machine. The Github is a cloud platform hosting Git repositories, enabling collaboration and remote storage.
We use Git locally to commit changes and push to a Github repository to be accessible to others. After that, they can pull from Github, make changes and push back.  
A Pull Request is a Github feature to ask for merging changes from one branch to another (e.g., main). A pull request includes codes, comments and automated checks. It used when new features need to be implemented to request team members to review them before merging.
```
